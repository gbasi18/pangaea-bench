```python
import torch
from types import SimpleNamespace
from pangaea.decoders.upernet_decfb import SegMTUPerNetDecFb

s = SimpleNamespace(temporal_pool="cum_mean", _pool_m=None, _pool_t=0)
seq = [[torch.randn(2, 8, 4, 4)] for _ in range(6)]     # 6 frames, 1 layer
for k, taps in enumerate(seq):
    out = SegMTUPerNetDecFb._pool_taps(s, taps)
    expected = torch.stack([seq[i][0] for i in range(k + 1)]).mean(0)
    assert torch.allclose(out[0], expected, atol=1e-6), f"wrong mean at t={k}"
print("mean math OK")
```
